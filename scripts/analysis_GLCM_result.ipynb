```python
# Load the CSVs for different percentages (if available)
path_susu_100 = "D:\SEMESTER5\PCD\proyek_image_processing_susu\output\glcm_results\susu_100_glcm_result.csv"  # This would be the path to the dataset
path_susu_90 = "D:\SEMESTER5\PCD\proyek_image_processing_susu\output\glcm_results\susu_90_glcm_result.csv"
path_susu_80 = "D:\SEMESTER5\PCD\proyek_image_processing_susu\output\glcm_results\susu_80_glcm_result.csv"
path_susu_70 = "D:\SEMESTER5\PCD\proyek_image_processing_susu\output\glcm_results\susu_70_glcm_result.csv"
path_susu_60 = "D:\SEMESTER5\PCD\proyek_image_processing_susu\output\glcm_results\susu_60_glcm_result.csv"

# Reading the data assuming they follow the same structure
df_susu_100 = pd.read_csv(path_susu_100)
df_susu_90 = pd.read_csv(path_susu_90)
df_susu_80 = pd.read_csv(path_susu_80)
df_susu_70 = pd.read_csv(path_susu_70)
df_susu_60 = pd.read_csv(path_susu_60)
```

```python
def process_dataframe(df):
    """Drop non-numeric columns and compute the mean of the remaining columns."""
    df_numeric = df.drop(columns=['filename'])  # Dropping the 'filename' column
    return df_numeric.mean()  # Returning the mean of each column

# Process all the sample dataframes
mean_100 = process_dataframe(df_susu_100)
mean_90 = process_dataframe(df_susu_90)
mean_80 = process_dataframe(df_susu_80)
mean_70 = process_dataframe(df_susu_70)
mean_60 = process_dataframe(df_susu_60)

# Combine all the means into a single dataframe
mean_df = pd.DataFrame({
    'susu_100%': mean_100,
    'susu_90%': mean_90,
    'susu_80%': mean_80,
    'susu_70%': mean_70,
    'susu_60%': mean_60
})

mean_df
```

```python
# mengambil standar deviasi dari masing-masing data
def process_dataframe2(df):
    """Drop non-numeric columns and compute the mean of the remaining columns."""
    df_numeric = df.drop(columns=['filename'])  # Dropping the 'filename' column
    return df_numeric.std()  # Returning the mean of each column

# Process all the sample dataframes
std_100 = process_dataframe2(df_susu_100)
std_90 = process_dataframe2(df_susu_90)
std_80 = process_dataframe2(df_susu_80)
std_70 = process_dataframe2(df_susu_70)
std_60 = process_dataframe2(df_susu_60)

# Combine all the means into a single dataframe
std_df = pd.DataFrame({
    'susu_100%': std_100,
    'susu_90%': std_90,
    'susu_80%': std_80,
    'susu_70%': std_70,
    'susu_60%': std_60
})
std_df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the top-level categories (grouped labels) and their respective subcategories
concentrations = ['Fresh (Susu 100%)', 'Campuran (Susu 90%)', 'Campuran (Susu 80%)', 'Campuran (Susu 70%)', 'Campuran (Susus 60%)']
channels = ['Blue', 'Green', 'Red']

# The number of channels per concentration level
n_channels = len(channels)

# Offset untuk setiap kelompok bar
offset = 0.2

# Flatten the x-axis labels to create a list of repeated color channels under each concentration
x_labels = [f"{channel}" for _ in concentrations for channel in channels]

# Create tick positions for both levels
tick_positions = np.arange(len(x_labels))

# Extract the feature data for each color channel and percentage
contrast_blue = mean_df.loc['contrast_B']
contrast_green = mean_df.loc['contrast_G']
contrast_red = mean_df.loc['contrast_R']

homogeneity_blue = mean_df.loc['homogeneity_B']
homogeneity_green = mean_df.loc['homogeneity_G']
homogeneity_red = mean_df.loc['homogeneity_R']

disimilarity_blue = mean_df.loc['dissimilarity_B']
disimilarity_green = mean_df.loc['dissimilarity_G']
disimilarity_red = mean_df.loc['dissimilarity_R']

energy_blue = mean_df.loc['energy_B']
energy_green = mean_df.loc['energy_G']
energy_red = mean_df.loc['energy_R']

correlation_blue = mean_df.loc['correlation_B']
correlation_green = mean_df.loc['correlation_G']
correlation_red = mean_df.loc['correlation_R']

entropy_blue = mean_df.loc['entropy_B']
entropy_green = mean_df.loc['entropy_G']
entropy_red = mean_df.loc['entropy_R']

# Combine the channels for Contrast and Homogeneity
contrast = [contrast_blue[0], contrast_green[0], contrast_red[0],
            contrast_blue[1], contrast_green[1], contrast_red[1],
            contrast_blue[2], contrast_green[2], contrast_red[2],
            contrast_blue[3], contrast_green[3], contrast_red[3],
            contrast_blue[4], contrast_green[4], contrast_red[4]]

homogeneity = [homogeneity_blue[0], homogeneity_green[0], homogeneity_red[0],
               homogeneity_blue[1], homogeneity_green[1], homogeneity_red[1],
               homogeneity_blue[2], homogeneity_green[2], homogeneity_red[2],
               homogeneity_blue[3], homogeneity_green[3], homogeneity_red[3],
               homogeneity_blue[4], homogeneity_green[4], homogeneity_red[4]]

disimilarity = [disimilarity_blue[0], disimilarity_green[0], disimilarity_red[0],
                disimilarity_blue[1], disimilarity_green[1], disimilarity_red[1],
                disimilarity_blue[2], disimilarity_green[2], disimilarity_red[2],
                disimilarity_blue[3], disimilarity_green[3], disimilarity_red[3],
                disimilarity_blue[4], disimilarity_green[4], disimilarity_red[4]]

energy = [energy_blue[0], energy_green[0], energy_red[0],
            energy_blue[1], energy_green[1], energy_red[1],
            energy_blue[2], energy_green[2], energy_red[2],
            energy_blue[3], energy_green[3], energy_red[3],
            energy_blue[4], energy_green[4], energy_red[4]]

correlation = [correlation_blue[0], correlation_green[0], correlation_red[0],
                correlation_blue[1], correlation_green[1], correlation_red[1],
                correlation_blue[2], correlation_green[2], correlation_red[2],
                correlation_blue[3], correlation_green[3], correlation_red[3],
                correlation_blue[4], correlation_green[4], correlation_red[4]]

entropy = [entropy_blue[0], entropy_green[0], entropy_red[0],
            entropy_blue[1], entropy_green[1], entropy_red[1],
            entropy_blue[2], entropy_green[2], entropy_red[2],
            entropy_blue[3], entropy_green[3], entropy_red[3],
            entropy_blue[4], entropy_green[4], entropy_red[4]]



```
